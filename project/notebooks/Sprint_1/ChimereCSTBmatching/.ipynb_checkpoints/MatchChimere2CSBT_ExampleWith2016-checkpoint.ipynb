```python
from netCDF4 import Dataset
import numpy as np
from mpl_toolkits.basemap import Basemap
```

## Chimère data

```python
# Open Chimère data
# here year : 2000
file = 'DailyGriddedData_2016.FRA03/INERIS.REANALYSED.FRA03.2016.O3.daymean.2gis.nc' # mention the path to the downloaded file
data = Dataset(file, mode='r') # read the data 
print(type(data)) # print the type of the data 
print(data.variables.keys()) # print the variables in the data
print(data.variables['Times'])
```

```python
lats = data.variables['lat'][:]  
longs = data.variables['lon'][:]
time = data.variables['Times'][:]
polution=data.variables['O3'][:] #O3 polution with time, lat, lon
```

```python
# Create basemap for visualization 
mp = Basemap(projection='merc',
             llcrnrlon=np.min(longs),   # lower longitude
             llcrnrlat=np.min(lats),    # lower latitude
             urcrnrlon=np.max(longs),   # uppper longitude
             urcrnrlat=np.max(lats),   # uppper latitude
            resolution = 'i')
```

```python
lon, lat = np.meshgrid(longs,lats)  #this converts coordinates into 2D arrray
x,y = mp(lon,lat) #mapping them together 
plt.figure(figsize=(16,8)) #figure size 
c_scheme = mp.pcolor(x,y,np.squeeze(polution[0,:,:]),cmap = 'jet') # [0,:,:] is for the first day of the year
plt.clim(0,150)

# consider this as the outline for the map that is to be created 
mp.drawcoastlines()
mp.drawstates()
mp.drawcountries()

cbar = mp.colorbar(c_scheme,location='right',pad = '10%') # map information
plt.title('Mean O3 concentration [ug/m3] : Day 1 of year 2016')


plt.show()
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    #print('value to find', value)
    #print('nereast value',array[idx])
    return idx,array[idx]
```

```python
polution_near_bati=[]
for i,L in enumerate(df_bati.Longitude):
    polution_near_bati.append(
                polution[0, #day1 of year chosen !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
                    find_nearest(lats,df_bati.Latitude.values[i])[0],
                    find_nearest(longs,L)[0]
                  ]
                 )
```

## Transformation to get number of days in the year = index in chimère data file 

```python
dt64 =df_bati.Date_debut_enquete.values[0]
print('Test transformation of datetime 64 to number of the day in the year')
print(dt64)

index_day = datetime.strptime(str(dt64),"%Y-%m-%dT%H:%M:%S.000000000").timetuple().tm_yday - 1 #python count from 0 
print(index_day)
print('verification ...')
print('    ', datetime.strptime(str(time.data[index_day])[:-10], '%Y%m%d'))
```

```python
def find_index_day(dt64):
    return datetime.strptime(str(dt64),"%Y-%m-%d %H:%M:%S").timetuple().tm_yday - 1 #python count from 0 
    
```

```python
df_bati.head()
```

## Loop on date (2016 for now !)


```python
polution_near_bati = []
lats_bati = []
longs_bati = []
date_debut=[]
date_fin=[]
iden_bati=[]


# set the index to column 'A'
#test = test.set_index('Identifiant')


for date in df_bati.Date_debut_enquete: 
    if (date.year == 2016):
        df =df_bati[df_bati.Date_debut_enquete==date]
        
        # select the row(s) where index value is 2

        for i,L in enumerate(df.Longitude):
            iden_bati.append(df.Identifiant.values[i])
            lats_bati.append(df.Latitude.values[i])
            longs_bati.append(L)
            date_debut.append(df.Date_debut_enquete.values[i])
            date_fin.append(df.Date_fin_enquete.values[i])


            polution_near_bati.append(
                             polution[find_index_day(date), 
                                find_nearest(lats,df.Latitude.values[i])[0],
                                find_nearest(longs,L)[0]
                              ]
                             )
```

```python
d = {'Identifiant':iden_bati,
     'Longitude':longs_bati,
     'Latitude':lats_bati,
     'Date_debut_enquete':date_debut,
     'Date_fin_enquete':date_fin,
     'O3meanChimere':polution_near_bati}
    
    
df = pd.DataFrame(d)
```
